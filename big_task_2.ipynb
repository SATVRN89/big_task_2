Let's work a bit with JSON. It's also a useful type of data, often used alongside dataframes and various tables.

The structure of the file is very simple: it shows which city can be reached from another city (where routes exist).

- First, display a list of all cities from which there are routes.
- Display a list of all cities to which there are routes.
- And as a bonus—calculate, for each departure city, the number of cities it is possible to travel to.

```python
import pandas as pd
import json
import warnings                   
warnings.filterwarnings('ignore')
```

```python
# Open json file
with open('C:/Users/ii/Desktop/SATVRN/study/Andrei/scoring_routes.json', encoding="utf8") as json_data:
    data = json.load(json_data)
```

```python
# Display a list of all cities from which there are routes
print(list(data.keys()))
```

```python
# Display a list of all cities to which there are routes sorted alphabetically
def flatten(xss):
    return [x for xs in xss for x in xs]
values_list = flatten(list(data.values()))
print(sorted(set(values_list)))
```

```python
# Calculate for each departure city the number of cities it is possible to travel to
city_counts = []
for k in list(data.keys()):
    city_counts.append(len(data[k]))
print(city_counts)
```

```python
# Create a dataframe with the city name and the number of routes for each city
df_1 = pd.DataFrame([city_counts], columns=list(data.keys()))
df_1
```

```python
# Create a dataframe with all the routes from this city
df_2 = pd.DataFrame(dict([(k, pd.Series(v)) for k, v in data.items()]))
df_2.tail(5)
```

```python
# Concatenate dataframes where the first row is a number of routes and then
# go all the routes for each city 
df = pd.concat([df_1, df_2], ignore_index=True, axis=0)
df = df.fillna('-')
df.head(10)
```
